# Introduction

Google Trends gives us an estimate of search volume. Let's explore if search popularity relates to other kinds of data. Perhaps there are patterns in Google's search volume and the price of Bitcoin or a hot stock like Tesla. Perhaps search volume for the term "Unemployment Benefits" can tell us something about the actual unemployment rate? 

Data Sources: <br>
<ul>
<li> <a href="https://fred.stlouisfed.org/series/UNRATE/">Unemployment Rate from FRED</a></li>

### Tesla Stock Price v.s. Search Volume

**Challenge:** Plot the Tesla stock price against the Tesla search volume using a line chart and two different axes. Label one axis 'TSLA Stock Price' and the other 'Search Trend'. 

```python
plt.rcParams.keys()
```

```python
years = mdates.YearLocator()
```

```python
months = mdates.MonthLocator()
```

```python
years_fmt = mdates.DateFormatter('%Y')
```

```python
# plt.figure(figsize=(14,8), dpi=120)
```

```python
plt.rcParams['figure.figsize'] = [14, 8]
```

```python
plt.rcParams['figure.dpi'] = 120
```

```python
fig, axis_1 = plt.subplots()
```

```python
axis_2 = axis_1.twinx()
```

```python
axis_1.xaxis.set_major_locator(years)
```

```python
axis_1.xaxis.set_major_formatter(years_fmt)
```

```python
axis_1.xaxis.set_minor_locator(months)
```

```python
axis_1.set_xlabel('DATE', fontsize=17)
```

```python
axis_1.set_ylabel('TSLA Stock Price', fontsize=20, color='#E6232E')
```

```python
axis_2.set_ylabel('Search Trend', fontsize=20, color='skyblue')
```

```python
# Set the minimum and maximum values on the axes - this generates ConversionError: Failed to convert value(s) to axis units: '2010-06-01' 
# axis_1.set_ylim([0, 600])
# axis_1.set_xlim([df_tesla.MONTH.min(), df_tesla.MONTH.max()])
```

```python
axis_1.plot(pd.to_datetime(df_tesla['MONTH']), df_tesla['TSLA_USD_CLOSE'], label='TSLA Stock Price', color='#E6232E', linewidth=3)
```

```python
axis_2.plot(pd.to_datetime(df_tesla['MONTH']), df_tesla['TSLA_WEB_SEARCH'], label='Search Trend', color='skyblue', linewidth=3) 
```

```python
axis_1.set_ylim(0, 600)
```

```python
axis_1.set_xlim(pd.to_datetime(df_tesla.MONTH.min()), pd.to_datetime(df_tesla.MONTH.max()))
```

```python
fig.legend(loc='upper left', borderpad=3, borderaxespad=12, title='Tesla Web Search vs Price', title_fontsize=15)
```

```python
for tick in axis_1.get_xticklabels():
    tick.set_rotation(45)
    tick.set_fontsize(14)
```

```python
for tick in axis_1.get_yticklabels():
    tick.set_fontsize(14)
    tick.set_color('#E6232E')
```

```python
for tick in axis_2.get_yticklabels():
    tick.set_fontsize(14)
    tick.set_color('skyblue')
```
